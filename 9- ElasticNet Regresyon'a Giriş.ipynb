Amacımız, hata kareler toplamını minimize eden katsayıları bu katsayılara bir ceza uygulayarak bulmaktır.

- ElasticNet L1 (Lasso Regresyon) ve L2 (Ridge Regresyon) yaklaşımlarını birleştirir.

- ElasticNet daha etkin bir düzgünleştirme işlemi yapar. Yani Ridge tarzı cezalandırma, Lasso tarzı değişken seçimi yapar.

- Çoklu doğrusal bağlantı problemi ve aşırı öğrenme problemlerine karşı daha dirençlidir.

- Formüldeki alfa parametreleri bu denklemler çözülüp açıldığında tek bir alfa değerine indirgenecek.
