- Lasso Regresyon ve Ridge Regresyon literatürde düzenlilieştirme yöntemleri olarak da geçmektedir.
    - Lasso regresyonu için "L1 yöntemi" ,
    - Ridge regresyonu için "L2 yöntemi" denir.



- Ridge regresyonunun ilgili-ilgisiz tüm değişkenleri modelde bırakma dezavantajını gidermek için önerilmiştir.

- Lasso da katsayıları sıfıra yaklaştırır. (Böylece değişken seçimi işlemini gerçekleştirmiş olur.) Fakat L1 normu, λ yeteri kadar büyük olduğunda bazı katsayıları sıfır yapar. Böylece değişken seçimi yapmış olur. (Ridge'de sıfıra yaklaşıyordu sıfır olmuyordu. Bu yüzden tüm değişkenler modelde kalıyordu.)    !!!! fark bu

- λ'nın doğru seçilmesi çok önemlidir. Burada da CV (cross validation) kullanılır.

- Ridge ve Lasso yöntemleri birbirinden üstün değildir.

# λ AYAR PARAMETRESİNİN BELİRLENMESİ

- λ'nın sıfır olduğu yer EKK'dir (klasik regresyondur). HKT'yi (hata kareler toplamı) minimum yapan λ'yı arıyoruz.

- λ için belirli değerleri içeren bir küme seçilir ve her birisi için cross validation test hatası hesaplanır.

- En küçük cross validation hata oranını veren λ ayar parametresi olarak seçilir.

- Son olarak seçilen bu λ ile model yeniden tüm gözlemlere fit edilir.
