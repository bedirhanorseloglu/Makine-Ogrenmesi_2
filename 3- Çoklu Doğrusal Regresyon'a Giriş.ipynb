# Doğrusal Regresyonun Varsayımları

- Hatalar normal dağılır.

- Hatalar birbirinden bağımsızdır ve aralarında otokorelasyon yoktur

- Her bir gözlem için hata terimleri varyansları sabittir.

- Değişlenler ile hata terimi arasında ilişki yoktur.

- Bağımsız değişkenler arasında çoklu doğrusal ilişki problemi yoktur.

Bunlar bir regresyon modelini klasik yöntemlerle kullanacak olduğumuzda  katsayıları (parametre tahmincilerini) tahmin edeceksek göz önünde bulundurduğumuz varsayımlardır. Ancak bu varsayımlar sağlandığında doğrusal modellerin sağlamış olduğu bazı neden-sonuç yorumlamalarını yapabiliyoruz.

# Regresyon Modellerinin Avantaj ve Dezavantajları

- İyi anlaşılırsa diğer tüm ML (machine learning) ve DL (deep learning) konuları çok rahat kavranır. (+)

- Doğrusallık nedensellik yorumları yapılabilmesini sağlar. Bu durum aksiyoner ve stratejik modelleme imkanı verir. (+)

- Değişkenlerin etki düzeyleri ve anlamlılıkları değerlendirilebilir. (+)

- Bağımlı değişkendeki değişklenliğin açıklanma başarısı ölçülebilir. (+) -> score

- Model anlamlılığı vardır. (+)

- Varsayımları vardır. (-)

- Aykırı gözlemlere duyarlıdır. (-)
